## Bag of Word


```python
import re

# Bài tập: Tạo Bag-Of-Word cho tập dataset sau: corpus = ["Tôi thích môn Toán", "Tôi thích AI", "Tôi thích âm nhạc"].
# Sau đó tạo list có tên vector để lưu vector sau khi thực hiện bước Tokenization đoạn văn bản sau
#  Tôi thích AI thích Toán, biết Bag-Of-Word được sắp theo thứ tự tăng dần
def to_word_dict(sentences):
  words=[
    word
    for wordSentence in [re.split(r'\s+', sentence) for sentence in sentences]
    for word in wordSentence
  ]

  word_dict = {}
  for word in words:
    word_dict[word] = word_dict.get(word, 0) + 1
  return word_dict


corpus = ["Tôi thích môn Toán", "Tôi thích AI", "Tôi thích âm nhạc"]

word_dict = to_word_dict(corpus)
bow = [item[0] for item in sorted(word_dict.items(), key=lambda item: item[0] )]

print(bow)

input = "Tôi thích AI thích Toán"
input_word_dict = to_word_dict([input])

print([input_word_dict.get(w, 0) for w in bow])
```

## Algorithm on List

```python
# • Tạo List có tên là lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]. Sau đó, áp
# dụng phương pháp tìm kiếm để tìm vị trí có giá trị None có trong lst_data theo 2 cách:
# tìm vị trí None đầu tiên, và tìm tất cả vị trí có giá trị None
#
# lst_data = [ ’’’ Your Code Here ’’’]
# # Your code here
# Output:
# - Vị trí None đầu tiên: 2 - Danh sách vị trí có giá trị None: [2, 4, 6]

lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]

# Solution 1:
for i in range(len(lst_data)):
  if lst_data[i] == None:
    print(f'Vị trí None đầu tiên: {i}')
    break

# Solution 2:
indexes = []
for i in range(len(lst_data)):
  if lst_data[i] == None:
    indexes.append(i)

print(f'Danh sách vị trí có giá trị None: {indexes}')


```

## Interpolation for List (Time-series)

```python
# • Tạo List có tên là lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]. Sau đó,
# áp dụng phương pháp nội suy Nearest Neighbor để gắn giá trị None có trong lst_data
#
# lst_data = [ ’’’ Your Code Here ’’’]
# # Your code here
# Output:
# - [1, 1.1, 1.1, 1.4, 1.4, 1.5, 1.5, 2.0]

lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]

# Neareast Neighbor
for i in range(len(lst_data)):
  if lst_data[i]:
    continue

  step = 1
  substitude_value = None
  while i - step >= 0 and i + step < len(lst_data):

    if lst_data[i - step]:
      lst_data[i] = lst_data[i - step]
      break;
    elif lst_data[i + step]:
      lst_data[i] = lst_data[i + step]
      break;

    step += 1


print(lst_data)
```

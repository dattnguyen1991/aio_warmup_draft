## Bag of Word


```python
import re

# Bi tp: To Bag-Of-Word cho tp dataset sau: corpus = ["Ti thch mn Ton", "Ti thch AI", "Ti thch m nhc"].
# Sau  to list c tn vector  lu vector sau khi thc hin bc Tokenization on vn bn sau
#  Ti thch AI thch Ton, bit Bag-Of-Word c sp theo th t tng dn
def to_word_dict(sentences):
  words=[
    word
    for wordSentence in [re.split(r'\s+', sentence) for sentence in sentences]
    for word in wordSentence
  ]

  word_dict = {}
  for word in words:
    word_dict[word] = word_dict.get(word, 0) + 1
  return word_dict


corpus = ["Ti thch mn Ton", "Ti thch AI", "Ti thch m nhc"]

word_dict = to_word_dict(corpus)
bow = [item[0] for item in sorted(word_dict.items(), key=lambda item: item[0] )]

print(bow)

input = "Ti thch AI thch Ton"
input_word_dict = to_word_dict([input])

print([input_word_dict.get(w, 0) for w in bow])
```

## Algorithm on List

```python
#  To List c tn l lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]. Sau , p
# dng phng php tm kim  tm v tr c gi tr None c trong lst_data theo 2 cch:
# tm v tr None u tin, v tm tt c v tr c gi tr None
#
# lst_data = [  Your Code Here ]
# # Your code here
# Output:
# - V tr None u tin: 2 - Danh sch v tr c gi tr None: [2, 4, 6]

lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]

# Solution 1:
for i in range(len(lst_data)):
  if lst_data[i] == None:
    print(f'V tr None u tin: {i}')
    break

# Solution 2:
indexes = []
for i in range(len(lst_data)):
  if lst_data[i] == None:
    indexes.append(i)

print(f'Danh sch v tr c gi tr None: {indexes}')


```

## Interpolation for List (Time-series)

```python
#  To List c tn l lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]. Sau ,
# p dng phng php ni suy Nearest Neighbor  gn gi tr None c trong lst_data
#
# lst_data = [  Your Code Here ]
# # Your code here
# Output:
# - [1, 1.1, 1.1, 1.4, 1.4, 1.5, 1.5, 2.0]

lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]

# Neareast Neighbor
for i in range(len(lst_data)):
  if lst_data[i]:
    continue

  step = 1
  substitude_value = None
  while i - step >= 0 and i + step < len(lst_data):

    if lst_data[i - step]:
      lst_data[i] = lst_data[i - step]
      break;
    elif lst_data[i + step]:
      lst_data[i] = lst_data[i + step]
      break;

    step += 1


print(lst_data)
```
